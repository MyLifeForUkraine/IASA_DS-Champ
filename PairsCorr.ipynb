```python
data_list = []
for row in corrmat.itertuples():
     for elem in row:
            print(elem)
            data_list.append(elem)
```

```python
pairs=[]
for i in range((len(data_list))):
    if(i%60==0):
        continue
    else:
        if(data_list[i]>0.95):
            temp=i//60
            pairs.append([metrix[temp+1],metrix[(i)%60]])
```
